```python
query = """
SELECT season, COUNT(DISTINCT stage) AS number_of_stages, AVG(home_team_goal) AS avg_home_team_goals, AVG(away_team_goal ) AS avg_away_team_goals, home_team_goal+away_team_goal AS total_goals
FROM Match
GROUP BY season;
"""

stages_by_season = pd.read_sql(query, conn)
stages_by_season
```

```python
query = """
SELECT c.name AS country_name, COUNT(m.id) AS count_matches
FROM Match m
JOIN Country c ON m.country_id = c.id
WHERE m.season = '2008/2009' AND c.name IN ('Spain', 'Belgium')
GROUP BY c.name
ORDER BY count_matches DESC;
"""

country_matches = pd.read_sql(query, conn)
country_matches
```

```python
query = """
SELECT m.id AS match_id, c.name AS country_name, l.name AS league_name, m.season, m.stage, m.date, ht.team_long_name AS home_team, at.team_long_name AS away_team, m.home_team_goal, m.away_team_goal
FROM Match m
JOIN Country c ON m.country_id = c.id
JOIN League l ON m.league_id = l.id
JOIN Team ht ON m.home_team_api_id = ht.team_api_id
JOIN Team at ON m.away_team_api_id = at.team_api_id
WHERE c.name = 'Spain'
ORDER BY m.date ASC
LIMIT 10;
"""

detailed_matches = pd.read_sql(query, conn)
detailed_matches
```

```python
query = """
SELECT
  c.name AS Country,
  l.name AS League,
  COUNT(DISTINCT m.stage) AS Stages,
  m.season AS Season,
  COUNT(DISTINCT ht.team_api_id) + COUNT(DISTINCT at.team_api_id) AS number_of_teams,
  ROUND(AVG(m.home_team_goal), 2) AS avg_local_goals,
  ROUND(AVG(m.away_team_goal), 2) AS avg_visitor_goals,
  ROUND((SUM(m.home_team_goal + m.away_team_goal) / COUNT(m.id)), 2) AS avg_goals,
  ROUND(AVG((m.home_team_goal - m.away_team_goal)), 2) AS avg_goal_dif,
  SUM(m.home_team_goal + m.away_team_goal) AS total_goals
FROM Match m
JOIN Country c ON m.country_id = c.id
JOIN League l ON m.league_id = l.id
JOIN Team ht ON m.home_team_api_id = ht.team_api_id
JOIN Team at ON m.away_team_api_id = at.team_api_id
WHERE c.name IN ('Spain', 'Germany', 'France', 'Italy', 'England')
GROUP BY c.name, l.name, m.season
HAVING ROUND((SUM(m.home_team_goal + m.away_team_goal) / COUNT(m.id)), 2) > 2.5;
"""

leages_by_season = pd.read_sql(query, conn)
leages_by_season
```

```python
#10.1.3
query = """
SELECT
  c.name AS Country,
  l.name AS League,
  ht.team_long_name AS Team,
  COUNT(m.stage) AS Stages,
  m.season AS Season,
  SUM(m.home_team_goal) AS sum_goals,
  ROUND(AVG(m.home_team_goal), 2) AS avg_goals
FROM Match m
JOIN Country c ON m.country_id = c.id
JOIN League l ON m.league_id = l.id
JOIN Team ht ON m.home_team_api_id = ht.team_api_id
WHERE c.name IN ('Spain', 'Germany', 'England')
GROUP BY c.name, l.name, ht.team_long_name, m.season
HAVING SUM(m.home_team_goal) > 50
ORDER BY m.season DESC, sum_goals DESC;
"""

teams_by_season = pd.read_sql(query, conn)
teams_by_season
```

```python
query = """
SELECT
  p.player_name,
  DATE(CURRENT_DATE) - p.birthday AS Age,
  pa.overall_rating,
  pa.potential,
  pa.date AS RatingDate
FROM Player p
JOIN Player_Attributes pa ON p.player_api_id = pa.player_api_id
WHERE p.player_name = 'Cristiano Ronaldo'
ORDER BY Age;
"""

ratingCristiano = pd.read_sql(query, conn)
ratingCristiano
```
