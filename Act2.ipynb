```python
#Importar libreria a implementar
import pandas as pd
```

```python
#Declarar df_1 con el archivo en excel a trabajar
df_1 = pd.read_excel('/content/example_ventas.xlsx')
```

```python
#Imprimir df_1
df_1
```

```python
#Ejercicio 1: Mostrar las primeras filas
df_1.head(5)
```

```python
#Calcular el total de ventas por region
ventas_por_region = df_1.groupby('Region')['Ventas'].sum()
print(ventas_por_region)
```

```python
#Filtrar el DF para mostrar solo las ventas de David
ventas_david = df_1[df_1['Vendedor']=='David']
print(ventas_david)
```

```python
#Calcular el promedio de ventas por region
ventas_por_producto = df_1.groupby('Region')['Ventas'].mean()
print(ventas_por_producto)
```

```python
#Calcular el promedio de ventas por region con dos decimales
ventas_por_producto = df_1.groupby('Region')['Ventas'].mean().round(2)
print(ventas_por_producto)
```

```python
#Ordenar el DF por ventas en orden descendente
df_ordenado = df_1.sort_values(by='Ventas', ascending=False)
print(df_ordenado)
```

```python
#FIltrar las filas con ventas mayores a 10000 y ordenar de menor a mayor
df_filtrado = df_1[df_1['Ventas']>10000].sort_values(by='Ventas', ascending=True)
print(df_filtrado)
```

```python
#Eliminar la columna "Vendedor"
df_sin_vendedor = df_1.drop(columns=['Vendedor'])
print(df_sin_vendedor)
```

```python
#Hacer una copia del DF original y agregar una columna de "Valor PesoCOL" con la conversion de euro a peso colombiano
euros = 4600
df_convertido = df_1.copy()
df_convertido['Valor PesoCOL'] = df_convertido['Ventas']*euros
df_convertido.head()
```

```python
#Calcular el total de ventas por producto
ventas_por_producto = df_1.groupby('Producto')['Ventas'].sum()
print(ventas_por_producto)
```

```python
#Calcular el total de vantas por vendedor
ventas_por_vendedor = df_1.groupby('Vendedor')['Ventas'].sum()
print(ventas_por_vendedor)
```
